# Problem statement:
Writing a python script which can fetch all the tweets(as many as allowed by Twitter API) done by ​ midas@IIITD​ twitter handle and dump the responses into JSONlines file.
The other part of the script should be able to parse these JSONline files to display the
following for every tweet in a tabular format.

## Necessary Imports 

```python
import tweepy #importing tweepy
import jsonlines # importing jsonlines
import json
import pandas as pd
import csv
```

## Defining Access tokens for Twitter API

```python
access_token = 'ENTER THE ACCESS TOKEN'
access_token_secret = 'ENTER THE ACCESS TOKEN SECRET'
access_key = 'ENTER YOUR API KEY'
access_secret = 'ENTER THE API KEY SECRET'
```

## Accessing All Timeline tweets by the user using Tweepy.Cursor

We use pagination a lot in Twitter API development. Iterating through timelines, user lists, direct messages, etc. In order to perform pagination we must supply a page/cursor parameter with each of our requests. The problem here is this requires a lot of boiler plate code just to mange the pagination loop. To help mage pagination easier and require less code Tweepy has the Cursor object.

```python
auth_tw = tweepy.OAuthHandler(access_key , access_secret) # Authentication of the API
auth_tw.set_access_token(access_token , access_token_secret)
api = tweepy.API(auth_tw)

all_tweets = []
for status in tweepy.Cursor(api.user_timeline, screen_name='@midasIIITD', tweet_mode="extended").items(): 
    all_tweets.append(status)               #appending all tweets in a list          
print(all_tweets[1])
```

## Saving the jsonline file of the tweets

```python
with open('sa.jsonl','a') as f:
    for t in all_tweets:
        json.dump(t._json,f)
        f.write('\n')
```

## Accessing the file and creating a tabular data

```python
final=[]; # A list to get all the tweets from output file
with jsonlines.open('sa.jsonl') as f:
    for i in f:
        final.append(i)
print((final[1]))
df = pd.DataFrame(columns=['full_text','created_at','favourite_count','retweet_count']) # DataFrame for the display

for i in range(len(final)):
    try:
        img = len(final[i]['extended_entities']['media'])  # image_count
        media_c = img

    except:
        media_c = None
    
        
    df = df.append({"full_text":final[i]['full_text'],
                     "created_at":final[i]['created_at'],
                     "favourite_count":final[i]['favorite_count'],
                     "retweet_count":final[i]['retweet_count'],
                     "image_count":media_c},ignore_index=True)
```
